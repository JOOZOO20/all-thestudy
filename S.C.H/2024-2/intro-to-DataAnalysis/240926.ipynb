```python
#1
mid['popkid']=mid['poptotal']-mid['popadults']
```

```python
mid['percentage'] = (mid['popkid']/mid['poptotal']) * 100
```

```python
#2
mid=mid.sort_values('percentage', ascending=False)
mid['county'].head()
```

```python
#3
import numpy as np

mid['grade']=np.where(mid['percentage']>=40, 'large',
                      np.where(mid['percentage']>=30,'middle','small'))
```

```python
mid.groupby('grade')\
    .agg(n=('grade','count'))
```

```python
#4
mid['per_a']=(mid['poptotal']/mid['popasian'])*100
mid=mid.sort_values('per_a')
mid.head()
```
