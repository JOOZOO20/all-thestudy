```python
# 막대그래프는 sns 의 barplot
```

```python
mpg.groupby('drv')\
.agg(mean_hwy=('hwy','mean'))
```

```python
df_mpg=mpg.groupby('drv')\
.agg(mean_hwy=('hwy','mean'))
```

```python
sns.barplot(data=df_mpg, x='drv',y='mean_hwy')
```

```python
# 높은 막대부터 나타내려면
df_mpg = df_mpg.sort_values('mean_hwy',ascending=False)
```

```python
df_mpg
```

```python
sns.barplot(data=df_mpg, x='drv',y='mean_hwy')
```

```python
df_mpg = mpg.groupby('drv')\
.agg(number=('drv','count'))
```

```python
df_mpg
```

```python
sns.barplot(data=df_mpg, x='drv', y='number')
```
