```python
# Ex 1-7
import numpy as np

# f(x)=x^3+1
def f(x):
    return x**3 + 1

# g(x)=sqrt(x+2)
def g(x):
    return np.sqrt(x+2)

# (g o f)(2), (f o g)(2), (g o g)(2)
print("(g o f)(2) =", g(f(2)))
print("(f o g)(2) =", f(g(2)))
print("(g o g)(2) =", g(g(2)))
```

```python
# Ex 1.14
import numpy as np
import matplotlib.pyplot as plt

# f(x)=3x-4
def f(x):
    return 3*x-4

# f^-1(x)=(x+4)/3
def f_inv(x):
    return (x+4)/3

# g(x)=x^2+1
def g(x):
    return x**2+1

# g^-1(x)=sqrt(x-1)
def g_inv(x):
    return np.sqrt(x-1)

# y = x
def identity(x):
    return x

# f = f^-1
x = np.linspace(-1,5,601)
graph, = plt.plot(x,f(x))
graph1, = plt.plot(x,f_inv(x))
graph2, = plt.plot(x,identity(x),'--')
plt.legend(handles=(graph,graph1,graph2), labels=(r'$y=f(x)$',r'$y=f^{-1}(x)$',r'$y=x$'))
plt.ylim(-1,5)
plt.show()

# g = g^-1
x = np.linspace(0,6,601)
graph, = plt.plot(x,g(x))
graph1, = plt.plot(x,g_inv(x))
graph2, = plt.plot(x,identity(x),'--')
plt.legend(handles=(graph,graph1,graph2), labels=(r'$y=g(x)$',r'$y=g^{-1}(x)$',r'$y=x$'))
plt.ylim(0,6)
plt.show()
```

```python
# Exercise 1-4
import numpy as np
import matplotlib.pyplot as plt

# (a) f(x)=x+2 (x<=-1), x^2 (x>-1)
x = np.linspace(-2,2,401)
fx = np.zeros_like(x)
for i in range(0,401):
	if x[i] <= -1:
		fx[i] = x[i]+2
	elif x[i] > -1:
		fx[i] = x[i]**2
plt.plot(x,fx)
plt.show()

# (b) g(x) = x+|x|
x = np.linspace(-2,2,401)
gx = x+np.abs(x)
plt.plot(x,gx)
plt.show()

```

```python
# Exercise 8-(C)
import numpy as np
import matplotlib.pyplot as plt

# 짝함수(x**2)와 홀함수(x**3) 정의
def even_function(x):
    return x**2

def odd_function(x):
    return x**3

# x 값 범위 설정
x = np.linspace(-10, 10, 400)

# y 값 계산
y_even = even_function(x)
y_odd = odd_function(x)

# 그래프 그리기
plt.figure(figsize=(12, 6))

# 짝함수 그래프
plt.subplot(1, 2, 1)
plt.plot(x, y_even, label='f(x) = x^2')
plt.title('짝함수 (Even Function)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

# 홀함수 그래프
plt.subplot(1, 2, 2)
plt.plot(x, y_odd, label='f(x) = x^3')
plt.title('홀함수 (Odd Function)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
