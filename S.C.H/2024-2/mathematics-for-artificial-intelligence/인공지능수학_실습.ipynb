```python
# 예재 7-8
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np

x = sym.Symbol('x')
f = 1/3*(x**3)+1/2*(x**2)+x+1

# 문제 (a)
diff_f = sym.diff(f,x)
print("(a) f의 도함수 = ", diff_f)
print('\n')

# 문제 (b)
diff2_f = sym.diff(diff_f,x)
print("(b) f의 이계도함수 = ", diff2_f)
print('\n')

# 문제 (c)
# x값 범위
x = np.arange(-5, 5, 0.1)

# 축 레이블 넣기
plt.xlabel('x axis')
plt.ylabel('y axis')

# 그리드
plt.grid(color = "black", alpha=0.5, linestyle='--')

# 그리고자 하는 식
plt.plot(x, 1/3*(x**3)+1/2*(x**2)+x+1, label='y = 1/3*($x^3$)+1/2*($x^2$)+x+1')
print("(c)")

# 범례
plt.legend()
plt.show()
```

```python
# 예제 7-12
import sympy as sym

# 문제 (a)
x = sym.Symbol('x')
h = (x**3+1)**20
diff_h = sym.diff(h,x)
print("(a) h의 도함수 = ", diff_h)

# 문제 (b)
# sym.cos() : 코사인함수
x = sym.Symbol('x')
h = sym.cos(x)**2
diff_h = sym.diff(h,x)
print("(b) h의 도함수 = ", diff_h)

# 문제 (c)
x = sym.Symbol('x')
h = sym.exp(x**2+1)
diff_h = sym.diff(h,x)
print("(c) h의 도함수 = ", diff_h)
```

```python
# 예제 7-16
import sympy as sym

x = sym.Symbol('x')
f = x**3-1
g = x**2-1
val = 1

# 함수 f(x), g(x)의 극한값 계산
lim_f = sym.limit(f, x, val)
lim_g = sym.limit(g, x, val)

# 로피탈 정리를 이용한 극한값 계산
diff_f = sym.diff(f,x)
diff_g = sym.diff(g,x)
LHospital_val = sym.limit(diff_f/diff_g, x, val)

print("극한값 = ", LHospital_val)
```

```python
# 예제 7-19
import sympy as sym

# sym.solve() : 방정식의 해를 구하는 함수
# sym.oo : ∞를 의미
x = sym.Symbol('x')
f = x**3 + x
diff_f = sym.diff(f,x)
print("f의 도함수 = ", diff_f)
print("f의 도함숫값이 0보다 큰 구간 = ", sym.solve([diff_f > 0, -sym.oo < x]))
print("\n")
```
